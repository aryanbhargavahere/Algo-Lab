```python
# Exercise 1
def func1(a, b):
  result = 0
  for i in range(0, b):
    result += a
  return result
```

```python
# Exercise 2
def func2(a, b):
  result = 1
  while b > 0:
    result = result * a
    b = b - 1
  return result
```

```python
# Sample function call: func2(2, 4)
x=print(func2(2,4))
#This function performs exponential power using repeated multiplication
start=time.time()
end=time.time()
x=end-start
print(f"Running Time: {x*1000}")
```

```python
# Exercise 3
def func3(num):
   result = 0
   while num > 0:
    result = result + num % 10
    num //= 10
    return result
```

```python
# Sample function call: func3(1234)
x=print(func3(1234))
#This function returns the last digit in a number
start=time.time()
end=time.time()
x=end-start
print(f"Running Time: {x*1000}")
```

```python
# Exercise 4
def func4(a, b):
    b.sort() # Time complexity for this function is O(Nlog N) where N = len(b)
    intersect = 0
    for x in a:
        if binary_search(b, x) >= 0:
            intersect = intersect + 1
    return intersect

def binary_search(arr, target):
    low, high = 0, len(arr) - 1
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1
    return -1
```

```python
# Sample function call: func4([1, 2, 3, 4], [2, 4, 6, 8])
x=print(func4([1,2,3,4],[2,4,6,8]))
#This funtion uses binary search to return the number of similar elements
start=time.time()
end=time.time()
x=end-start
print(f"Running Time: {x*1000}")
```

```python
# Exercise 5
def func5(n):
    guess = 1
    while guess * guess <= n:
        if guess * guess == n:
            return guess
        guess = guess + 1
    return None
```

```python
# Sample function call: func5(16)
x=print(func5(16))
#This function checks that whether a nummber is a perfect square or not and gives the square root
start=time.time()
end=time.time()
x=end-start
print(f"Running Time: {x*1000}")
```

```python
# Exercise 6
def func8(A):
    func8_helper(A, 0, len(A) -1)

def func8_helper(A, i, j):
    #Input: An array A and two indices i and j.
    n = j - 1 + 1
    if n ==2:
        if A[i] > A[j]:
            # Swap A[i] and A[j]
            A[i], A[j] = A[j], A[i]
        elif n > 2:
            m = n // 3
            func8_helper(A, i, j - m)
            func8_helper(A, i + m, j)
            func8_helper(A, i, j - m)
        return A
```

```python
#Sample function call: func8([3,1,2,7,8,6])
x=print(func8([3,1,2,7,8,6]))
#This function sorts the numbers in an array using the logic of merge sort
start=time.time()
end=time.time()
x=end-start
print(f"Running Time: {x*1000}")
```

```python
# Exercise 7
def func12(arr, k):
    arr_len = len(arr)
    k = k % arr_len
    # Initialize a new array of the same length as 'arr'
    custom_arr = [0] * arr_len

    for i in range(arr_len):
        mystery_index = (i + k) % arr_len
        custom_arr[mystery_index] = arr[i]
    return custom_arr

```

```python
# write a function named as sum_arr to find sum of elements in given array
def sum_arr(arr):
  sum=0
  for i in arr:
    sum+=i
  return sum
```

```python
#write a function named as max_min to find minimum (or maximum) element in an array
MAX=0
MIN=0
def max_min(arr):
  for i in range(0,arr.size()):
    if MAX<arr[i]:
      MAX=arr[i]
    elif MIN>arr[i]:
      MIN=arr[i]
  return MAX,MIN
```

```python
# write a function named as Counting_EvenOdd to count number of even and odd elements in an array
def Counting_EvenOdd(arr):
  even=0
  odd=0
  for i in arr:
    if i%2==0:
      even+=1
    else:
      odd+=1
  return even,odd
```

```python
# write a function named as linear_Search to perform a linear search in 1D-array
def linear_Search(arr,key):
  arr_len=len(arr)
  for i in range(arr_len):
    if arr[i] == key:
      return i
```
